```python
import pandas as pd
import seaborn as sns
import plotly.express as px
from matplotlib import pyplot as plt
import numpy as np
```

```python
#df=dataframe sample_data是google內建測試資料夾
df=pd.read_csv('sample_data/california_housing_test.csv')
```

```python
df #橫:row（列）  直:column（欄）
```

```python
#散點圖
plt.subplot(2,2,1) #圖的位置
sns.scatterplot(df,x='median_income',y='median_house_value')

#回歸直線(regression)
plt.subplot(2,2,2)#圖的位置
sns.regplot(df,x='median_income',y='median_house_value')

#直方圖(histogram)
plt.subplot(2,2,3)#圖的位置
sns.histplot(df,x='housing_median_age')

#經緯度
plt.subplot(2,2,4) #圖的位置
sns.scatterplot(df,x='longitude',y='latitude')

plt.show()
```

```python
fig=px.scatter_map(df,lat='latitude',lon='longitude',color='median_house_value',
                   size='population',zoom=5)
fig.show()
```
